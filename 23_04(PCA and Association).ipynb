```python
import pandas as pd
from mlxtend.preprocessing import TransactionEncoder
```

```python
dataset = [['Milk','Onion','Nutmeg','KB','Mushrrom','yogurt'],
           ['Dell','Onion','Nutmeg','KB','Mushrrom','yogurt'],
           ['Milk','Apple','KB','Mushrrom'],
           ['Milk','Unicorn','Corn','KB','yogurt'],
           ['Corn','Onion','Onion','KB','Mushrrom','Ice cream']]
```

```python
te = TransactionEncoder()
te_ary = te.fit(dataset).transform(dataset)
```

```python
df = pd.DataFrame(te_ary,columns=te.columns_)
```

```python
############################### Apriori ###################

from mlxtend.frequent_patterns import apriori
```

```python
ar = apriori(df,min_support=0.6,use_colnames=True)
```

```python
######################## Association rule Mining ###########

from mlxtend.frequent_patterns import association_rules

asr = association_rules(ar,metric = 'confidence',min_threshold =  0.7)
```
